```python
import pandas as pd
import plotly.express as px

from notnews import *
```

```python
big_dd = pd.DataFrame()
```

```python
for i in range(6):
    path = f'../dataverse_files/cnn-{i+1}.csv'
    print (path)
    df = pd.read_csv(path, encoding = 'ISO-8859-1')
    df['channel.name'] = 'CNN'
    soft_news_us_df = pred_soft_news_us(df, col='text')
    big_dd = pd.concat([big_dd,soft_news_us_df], axis=0, ignore_index=True)
```

```python
soft_news_df_grp = big_dd.groupby(['program.name','channel.name'])['prob_soft_news_us'].mean().reset_index()
```

```python
soft_news_df_grp
```

```python
# Use column names of df for the different parameters x, y, color, ...
fig = px.scatter(soft_news_df_grp, x="prob_soft_news_us", y="program.name", color="channel.name",
                 title="Program by Channel distribution",
                 labels={"date"} # customize axis label
                )

fig.show()
```
