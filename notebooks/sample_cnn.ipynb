```python
big_dd = pd.read_csv('../dataverse_files/all_cnn_soft_news.csv')
```

```python
sample_df = big_dd[big_dd['channel.name'] == 'CNN'].sample(frac=0.1, random_state=10)
```

```python
sample_df
```

```python
def split_text(s, n):
    pieces = str(s).split()
    return (' '.join(pieces[i:i+n]) for i in range(0, len(pieces), n))
```

```python
for i, row in sample_df.iterrows():
    w_generator = split_text(row['text'],500)
    for j, word in enumerate(w_generator):
        col_name = 'text' + str(j)
        sample_df.loc[i,col_name] = word
```

```python
sample_df.to_csv('../dataverse_files/random_CNN_parsed.csv')
```
